```python
# data preprocessing
def replacer(A):
    import pandas as pd
    Q=pd.DataFrame(A.isna().sum(),columns=["pp"])
    for i in Q[Q.pp>0].index:
        if (A[i].dtypes=="object"):
            x=A[i].mode()
            A[i]=A[i].fillna(x)
        else:
            x=A[i].mean()
            A[i]=A[i].fillna(x)
            
            
```
